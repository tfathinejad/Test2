```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
city_csv = os.path.join("raw_data","city_data.csv")
city_df = pd.read_csv(city_csv)
city_df.head()
```

```python
ride_csv = os.path.join("raw_data","ride_data.csv")
ride_df = pd.read_csv(ride_csv)
ride_df.head()
```

```python
#merge
merged_df = pd.merge(city_df,ride_df, on='city', how='left')
merged_df.head()
```

```python
grouped_merged = merged_df.groupby(["type"])
grouped_merged.head(3)
```

```python
# create 3 sub data frame that you can filter urben suburben rural 
```

```python
rural_df = merged_df.loc[merged_df["type"]== "Rural"]
rural_df.head()
```

```python
rural_grouped = rural_df.groupby(["city"])
rural_grouped.head(2)
```

```python
urban_df= merged_df.loc[merged_df["type"]=="Urban"]
urban_df.head()
```

```python
urban_grouped=urban_df.groupby(["city"])
urban_grouped.head(1)
```

```python
suburban_df = merged_df.loc[merged_df["type"]=="Suburban"]
suburban_df.head()
```

```python
suburban_grouped=suburban_df.groupby(["city"])
suburban_grouped.head(1)
```

```python
# 1. Total number of rides per city
rural_rides = rural_grouped["ride_id"].count()

urban_rides = urban_grouped["ride_id"].count()

suburban_rides = suburban_grouped["ride_id"].count()
rural_rides
```

```python
# 2. Average fare per city

rural_fare = rural_grouped["fare"].mean()

urban_fare = urban_grouped["fare"].mean()

suburban_fare = suburban_grouped["fare"].mean()
```

```python
#Getting rid of duplicates in city, to count accurate number of drivers
rural_df_no_dup = rural_df.drop_duplicates(["city"])
# rural_df_no_dup
urban_df_no_dup = urban_df.drop_duplicates(["city"])
suburban_df_no_dup = suburban_df.drop_duplicates(["city"])
```

```python
# This is the groupby version with no duplicates, purpose is to count number of drivers in each city:
rural_grouped_no_dup= rural_df_no_dup.groupby(["city"])
urban_grouped_no_dup = urban_df_no_dup.groupby(["city"])
suburban_grouped_no_dup = suburban_df_no_dup.groupby(["city"])
```

```python
# 3. Number of drivers per city
# I realized using regular grouped values is wrong since it counts drivers for each city multiple times.
# Therefore removed duplicates from original subdata frames, then grouped by city.
# Please comment if this assumption is incorrect:

rural_driver_count = rural_grouped_no_dup["driver_count"].sum()
urban_driver_count = urban_grouped_no_dup["driver_count"].sum()
suburban_driver_count = suburban_grouped_no_dup["driver_count"].sum()

```

```python
# plt.scatter(temp, sales, marker='*', facecolors='yellow', edgecolors='blue')
plt.scatter(rural_rides,rural_fare,rural_driver_count, c="red", alpha=20, edgecolors="red", linewidth=2)
plt.scatter(urban_rides,urban_fare, urban_driver_count, c="green", alpha=0.6, edgecolors ="black", linewidth=2)
plt.scatter(suburban_rides, suburban_fare,suburban_driver_count, c="yellow", alpha=0.6, edgecolors = "black", linewidth=2)
plt.xlabel("Total Number of Rides per City")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Rides Sharing Data 2016")
plt.grid()
plt.show()

# Was not able to add the appropriate legend:
# plt.legend(handles=["Rural", "Urban", "Suburban"], loc="best")
```

```python
# % of Total Fares by City Type
# % of Total Rides by City Type
# % of Total Drivers by City Type
```

```python
# To Calculate %Total Fares
total_rural_fares = rural_df["fare"].sum()
total_urban_fares = urban_df["fare"].sum()
total_suburban_fares = suburban_df["fare"].sum()
total_fares = merged_df["fare"].sum()

percent_rural_fares= (total_rural_fares/total_fares)*100
percent_urban_fares= (total_urban_fares/total_fares)*100
percent_suburban_fares= (total_suburban_fares/total_fares)*100

# Just to remember how to do formatting on numbers:
# percent_rural_fares='%{:,.2f}'.format((total_rural_fares/total_fares)*100)
```

```python
labels = ["Rural","Urban","Suburban"]
percent = [percent_rural_fares,percent_urban_fares,percent_suburban_fares]
colors = ["pink", "gray", "yellow"]
explode = (0.1, 0, 0.1)
plt.pie(percent, explode=explode, labels=labels, colors=colors, autopct="%1.2f%%"
        , startangle=225)
plt.axis("equal")
plt.title("% of Total Fares by City Type")
plt.show()
```

```python
#To Calculate %Total Rides:

total_rides = grouped_merged["ride_id"].count()
total_rides.plot(kind='pie', colors=["pink","yellow","gray"], title= "% of Total Rides by City Type",autopct="%1.2f%%", figsize=(5,5))


plt.show()

```

```python
# To Calculate %Total Drivers
total_rural_drivers = rural_driver_count = rural_df_no_dup["driver_count"].sum()
total_urban_drivers = urban_driver_count = urban_df_no_dup["driver_count"].sum()
total_suburban_drivers = suburban_driver_count = suburban_df_no_dup["driver_count"].sum()
total_drivers = total_rural_drivers + total_urban_drivers + total_suburban_drivers
percent_rural_drivers = (total_rural_drivers/total_drivers)*100
percent_urban_drivers = (total_urban_drivers/total_drivers)*100
percent_suburban_drivers =(total_suburban_drivers/total_drivers)*100
```

```python

labels = ["Rural","Urban","Suburban"]
percent = [percent_rural_drivers,percent_urban_drivers,percent_suburban_drivers]
colors = ["pink", "yellow", "gray"]
explode = (0.1, 0, 0.1)
plt.pie(percent, explode=explode, labels=labels, colors=colors, autopct="%1.2f%%"
        , startangle=225)
plt.axis("equal")
plt.title("% of Total Drivers by City Type")
plt.show()
```
